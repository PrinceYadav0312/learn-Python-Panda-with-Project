# Pivot Table

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice'],
}
df = pd.DataFrame(data)
df
```

```python
# here we found region by product sales data.
df.pivot_table(values= "Sales", index="Region", columns="Product")
# Here by default it's give mean of sales.
# If want to change aggregation function then use 'aggfunc=' 
```

```python
pivot2 = df.pivot_table(values=['Sales', 'Units'], index="Region", columns="Product")
pivot2   # Make Sure in values we put only numeric value else if we give string value it's give error.
```
