# Merging Joining & Cocationation

```python
import numpy as np
import pandas as pd
```

Merging 2 Data Frames

```python
employee = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter',  'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DF2 Salary Information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employee
```

```python
salaries
```

```python
pd.merge(employee, salaries, on= 'employee_id', how='inner') # Here in on we have to give a common id for accurate merging.
# Here in how (inner) is bydefault it's merge on the bases of common value.
```

```python
pd.merge(employee, salaries, on= 'employee_id', how='outer') # here Outer merging on the bases of common column 
```

```python
# Here we use left when we want to get data only based on left side like in our case employee is in left side.
pd.merge(employee, salaries, on= 'employee_id', how='left') # So it merge on the base of employees. 
```

```python
# IF we want to get data on bases of right hand side data then we use 'Right'
pd.merge(employee, salaries, on= 'employee_id', how='right')
```

Concatination of 2 Data Frame

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2'],
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5'],
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # concat function only takes 1 parameter.
```

```python
pd.concat([df1,df2], axis= 1)
```

Joining of 2 DataFrames

```python
df3 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index= [1,2,3])

df4 = pd.DataFrame({
    'Score': [85, 90, 75]
}, index= [2, 3, 4])
```

```python
df3
```

```python
df4
```

```python
df3.join(df4, how= 'outer')
```

```python
df4.join(df3)
```
